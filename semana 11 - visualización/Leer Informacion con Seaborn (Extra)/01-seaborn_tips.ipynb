# Visualización de datos en Python con Seaborn

## Importación de Datos
Importamos la funcion load_dataset, para cargar data.


```python
from seaborn import load_dataset

tips = load_dataset('tips')
tips.head() # muestra los cinco primeros registros
```

## Gráficos de dispersion

Este gráfico permite ver la relación entre dos variables. Se usa el método implot, el cual acepta cómo argumento, la característica para cada uno de los ejes y el conjunto de datos.

```python
from seaborn import lmplot
lmplot('total_bill', 'tip', data=tips, fit_reg=False)
```

Usamos la opción *fig_reg = False*, dado que por defecto realiza regresión lineal.

```python
lmplot('total_bill', 'tip', data=tips)
```

```python
lmplot('total_bill', 'tip', data=tips, ci=None)
```

```python
lmplot(x='total_bill', y='tip', hue='time', data=tips, ci=None)
```

## Diagramas de densidad

```python
from seaborn import kdeplot
kdeplot(tips.total_bill)
```

```python
kdeplot(tips.total_bill, tips.tip)
```

## Histogramas

```python
from seaborn import distplot 
distplot(tips.total_bill)
```

```python
distplot(tips.total_bill, kde=False) #sin diagrama de densidad
```

```python
distplot(tips.total_bill, kde=False, hist=False)
```

```python
distplot(tips.total_bill, rug=True, hist=False)
```

```python
distplot(tips.total_bill, vertical=True)
```

## BoxPlot

```python
from seaborn import boxplot 

boxplot(tips.total_bill)
```

```python
boxplot(y=tips.total_bill, orient="v")
```

```python
boxplot(x='sex', y='total_bill', data=tips)
```

```python
boxplot(x='sex', y='total_bill', hue='smoker', data=tips)
```

```python
from seaborn import violinplot

violinplot(x='sex', y='total_bill', hue='smoker', data=tips)
```
